**Dataset tentang tingkat inflasi tahunan di Indonesia dari tahun 2000 hingga 2022. Data ini bisa diperoleh dari Bank Indonesia atau Badan Pusat Statistik.**

Fungsi utama dari visualisasi ini adalah untuk mentransformasikan data numerik tentang tingkat inflasi menjadi representasi visual yang mudah dipahami dan diinterpretasikan. Grafik ini berfungsi sebagai alat analitis yang memungkinkan pengguna untuk dengan cepat mengidentifikasi tren jangka panjang, fluktuasi jangka pendek, dan anomali dalam data inflasi. Selain itu, grafik ini berfungsi sebagai alat komunikasi yang efektif untuk menyampaikan informasi ekonomi kompleks kepada berbagai pemangku kepentingan, mulai dari pembuat kebijakan hingga masyarakat umum.

Tone yang digunakan dalam visualisasi ini dirancang untuk menjadi informatif, objektif, dan profesional. Meskipun demikian, visualisasi ini juga mempertahankan tingkat aksesibilitas dengan menggunakan elemen desain yang jelas dan mudah dibaca, seperti anotasi untuk poin-poin penting dan grid yang membantu interpretasi nilai. Keseimbangan antara kedalaman informasi dan kejelasan presentasi menciptakan tone yang mengundang pengguna untuk terlibat dengan data secara kritis, sambil tetap memastikan bahwa informasi dapat diakses oleh pengguna dengan berbagai tingkat keahlian ekonomi.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data inflasi tahunan Indonesia (2000-2022)
years = list(range(2000, 2023))
inflation_rates = [9.35, 12.55, 10.03, 5.06, 6.40, 17.11, 6.60, 6.59, 11.06, 2.78, 6.96, 3.79, 4.30, 8.38, 8.36, 3.35, 3.02, 3.61, 3.13, 2.72, 1.68, 1.87, 4.35]

# Membuat plot
plt.figure(figsize=(12, 6))
plt.plot(years, inflation_rates, marker='o')

# Menambahkan judul dan label
plt.title('Tingkat Inflasi Tahunan Indonesia (2000-2022)', fontsize=16)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Tingkat Inflasi (%)', fontsize=12)

# Menyesuaikan sumbu x
plt.xticks(years[::2], rotation=45)

# Menambahkan grid
plt.grid(True, linestyle='--', alpha=0.7)

# Menambahkan anotasi untuk nilai tertinggi dan terendah
max_inflation = max(inflation_rates)
min_inflation = min(inflation_rates)
max_year = years[inflation_rates.index(max_inflation)]
min_year = years[inflation_rates.index(min_inflation)]

plt.annotate(f'Tertinggi: {max_inflation}%', xy=(max_year, max_inflation), xytext=(5, 10),
             textcoords='offset points', ha='left', va='bottom',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
             arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

plt.annotate(f'Terendah: {min_inflation}%', xy=(min_year, min_inflation), xytext=(5, -10),
             textcoords='offset points', ha='left', va='top',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
             arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

# Menampilkan plot
plt.tight_layout()
plt.show()
```
